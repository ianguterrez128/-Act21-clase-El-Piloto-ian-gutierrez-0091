```python
class Piloto:
    def __init__(self, escuderia, auto, contrato):
        self.escuderia = escuderia
        self.auto = auto
        self.contrato = contrato
        self.motor_encendido = False

    def encender_motor(self):
        self.motor_encendido = True
        print(f"El motor del {self.auto} se ha encendido. ¡Vrum!")

    def mostrar_info(self):
        print(f"Piloto de: {self.escuderia}, Auto: {self.auto}, Contrato: ${self.contrato} MDD")

# creacion de objetos (instancias)
piloto1 = Piloto("Red Bull", "RB20", 50)
piloto2 = Piloto("Mercedes", "W15", 45)

# llamadas a las funciones para cada objeto piloto 1 y piloto 2
print("piloto 1:")
piloto1.mostrar_info()
piloto1.encender_motor()

print("\npiloto 2:")
piloto2.mostrar_info()
piloto2.encender_motor()

print("\nprograma terminada")
print("creador de software por ian gutierrez NL= 22")
```
