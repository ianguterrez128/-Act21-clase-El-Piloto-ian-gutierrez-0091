```python
class Piloto:
    def __init__(self, nombre, aeronave, horas_vuelo):
        self.nombre = nombre
        self.aeronave = aeronave
        self.horas_vuelo = horas_vuelo
        self.turbinas_encendidas = False

    def encender_motor(self):
        self.turbinas_encendidas = True
        print(f"Las turbinas del {self.aeronave} se han encendido. ¡Huuuum!")

    def mostrar_info(self):
        print(f"Capitán: {self.nombre}, Aeronave: {self.aeronave}, Experiencia: {self.horas_vuelo} Horas")

# Creación de objetos (instancias)
piloto1 = Piloto("Ian Gutierrez", "Boeing 787", 5000)
piloto2 = Piloto("Carlos Vazquez", "Airbus A320", 3500)

# Llamadas a las funciones para cada objeto
print("Piloto 1:")
piloto1.mostrar_info()
piloto1.encender_motor()

print("\nPiloto 2:")
piloto2.mostrar_info()
piloto2.encender_motor()

print("\nPrograma terminado")
print("Creador de software por Ian Gutierrez NL= 22")
```
